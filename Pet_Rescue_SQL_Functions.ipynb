## Step 1: Create and Populate PETRESCUE Table

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("pet_rescue.db")
cur = conn.cursor()

with open("PETRESCUE-CREATE.sql", "r") as file:
    sql_script = file.read()

# Fix the first line for SQLite
sql_script = sql_script.replace("drop table PETRESCUE;", "DROP TABLE IF EXISTS PETRESCUE;")

cur.executescript(sql_script)
conn.commit()

print("Table and data created successfully")
```

## Step 2: Verify Table Data

```python
pd.read_sql_query("SELECT * FROM PETRESCUE;", conn)
```

## Aggregate Functions

```python
pd.read_sql_query("""
SELECT SUM(COST)
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT SUM(COST)
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT SUM(COST) AS SUM_OF_COST
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT MAX(QUANTITY)
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT AVG(COST)
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT AVG(COST / QUANTITY)
FROM PETRESCUE
WHERE ANIMAL = 'Dog';
""", conn)
```

## Scalar and String Functions

```python
pd.read_sql_query("""
SELECT ROUND(COST)
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT LENGTH(ANIMAL)
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT UPPER(ANIMAL)
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT DISTINCT UPPER(ANIMAL)
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT *
FROM PETRESCUE
WHERE LOWER(ANIMAL) = 'cat';
""", conn)
```

## Date and Time Functions

```python
pd.read_sql_query("""
SELECT STRFTIME('%d', RESCUEDATE)
FROM PETRESCUE
WHERE ANIMAL = 'Cat';
""", conn)
```

```python
pd.read_sql_query("""
SELECT SUM(QUANTITY)
FROM PETRESCUE
WHERE STRFTIME('%m', RESCUEDATE) = '05';
""", conn)
```

```python
pd.read_sql_query("""
SELECT SUM(QUANTITY)
FROM PETRESCUE
WHERE STRFTIME('%d', RESCUEDATE) = '14';
""", conn)
```

```python
pd.read_sql_query("""
SELECT DATE(RESCUEDATE, '+3 day')
FROM PETRESCUE;
""", conn)
```

```python
pd.read_sql_query("""
SELECT JULIANDAY('now') - JULIANDAY(RESCUEDATE)
FROM PETRESCUE;
""", conn)
```
